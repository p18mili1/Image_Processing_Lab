```python
import numpy as np
from skimage import io
from skimage import data
import matplotlib.pyplot as plt

a = io.imread('parrot.png')
t = np.array([[1,0.9,0],[0,2,0],[0,0,1]])
lx,ly = a.shape
print(lx,ly)
plt.imshow(a , cmap='gray', vmin=0, vmax=255)
plt.savefig('parrot_1.png')
b = np.zeros((lx*2,ly*2), dtype=np.int16 )
b.shape
b = b + 255
print(b[0][0])
```

```python
for x in range(lx):
    for y in range(ly):
        axy = np.array([[x],[y],[1]])
        bxy = t @ axy
#        print(bxy[0][0],bxy[1][0])
        b[int(bxy[0][0]),int(bxy[1][0])] = a[x][y]
plt.imshow(b , cmap='gray', vmin=0, vmax=255)
#plt.savefig('parrot_05.png')
plt.savefig('parrot_x2.png',bbox_inches='tight', pad_inches=0)
```
